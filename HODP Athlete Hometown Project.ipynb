```python
import numpy as np
import pandas

import ssl
import certifi
import geopy.geocoders
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx

import geopy as gp
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

geolocator = Nominatim(user_agent="hodp")
```

```python
df = pandas.read_csv('HODP Athlete Project - All Athletes.csv', index_col = 0)
df = df[pandas.notnull(df['Hometown'])]
df = df.reset_index()
df.drop(columns = ['Unnamed: 8'])
df.head()
```

```python
def do_geocode(address):
    try:
        return geolocator.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)
```

```python
# Cleaner function
def cleaner(old, new):
    lst = [] 
    for i, row in df.iterrows():
        if old in row["Hometown"]:
            lst.append(i)

    print(lst)
    for i in lst:
        df.iloc[i]["Hometown"] = df.iloc[i]["Hometown"].replace(old, new)

cleaner("Ill.", "Illinois")
cleaner("Neb.", "Nebraska")
cleaner("Wisc.", "Wisconsin")
cleaner("Fla", "Florida")
cleaner("Coquitlan", "Coquitlam")
cleaner("Pozos de Santa", "Pozos")
cleaner("Cutchouge", "Cutchogue")
cleaner("Edgarton", "Edgartown")
```

```python
def do_geocode(address):
    try:
        return geolocator.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)
```

```python
locations = []

for index, row in my_df.iterrows():
    location = do_geocode(row["Hometown"])
    print(index, location)
#    my_df.loc[index, 'CHometown'] = location.address
#    print(index)
#    print(location.address)
    if location:
        locations.append(location)
    else:
        locations.append("")

my_df['CHometown'] = locations
```
